```python
import requests as req , re , pandas as pd
from bs4 import BeautifulSoup
```

```python
def channel_connection(channel_name):
    #channel_name = input('Enter Channel Name:')
    url = 'https://t.me/s/' + channel_name
    request = req.get(url)
    html = BeautifulSoup(request.content , 'html.parser')
    return html
```

```python
def convert_str_to_number(x):
    total_stars = 0
    num_map = {'K':1000, 'M':1000000, 'B':1000000000}
    if x.isdigit():
        total_stars = int(x)
    else:
        if len(x) > 1:
            total_stars = float(x[:-1]) * num_map.get(x[-1].upper(), 1)
    return int(total_stars)
```

```python
def subs (html , channel_name):
    counter = html.find_all('span', {'class' : 'counter_value'})[0]
    return convert_str_to_number(counter.get_text())
```

```python
def mentions(html , channel_name):
    relations_list_edge = []
    relations_list_node = []
    for a in html.find_all('a' , href = True):
        if re.search(r'https://t.me/[^[^/]+$' , a['href']) != None:
            link = a['href']
            suffix = link.rsplit('/' , 1)[-1]
            prefix = link.rsplit('/' , 1)[-2]
            new_link = prefix + '/s/' + suffix
            try:
              if channel_name.lower() !=suffix.lower() and suffix.lower() not in relations_list_node:
                  relations_list_node.append([suffix, subs(channel_connection(suffix), suffix)])
                  relations_list_edge.append([channel_name , suffix])
            except:
              if channel_name.lower() !=suffix.lower():                  
                relations_list_node.append([suffix, 'user and not a channel'])
                relations_list_edge.append([channel_name , suffix])

    
    node = pd.DataFrame(relations_list_node , columns=['Target', 'Subs'])
    edge = pd.DataFrame(relations_list_edge, columns=['Source', 'Target'])

    return edge, node
```
